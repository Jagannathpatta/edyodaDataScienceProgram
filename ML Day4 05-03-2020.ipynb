# Unsupervised Learning

```python
import pandas as pd
```

```python
house_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt',index_col='Unnamed: 0')
```

```python
house_data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_algo = KMeans(n_clusters=4)
```

```python
cluster_algo.fit(house_data)
```

```python
labels = cluster_algo.predict(house_data)
```

```python
labels
```

```python
house_data[labels == 3].head()
```

```python
house_data_subset = house_data[['Sqft','Price']]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(house_data_subset)
```

```python
labels = kmeans.predict(house_data_subset)
```

```python
plt.scatter(x=house_data_subset.Price, y=house_data_subset.Sqft, s=10, c=labels)
plt.xlabel('Price')
plt.ylabel('Sqft')
```

# Exercise
1. Find 3 categories of people in the company for HR_data
2. Steps
   * Transform all the categorical data to num
   * Use KMeans to find similar people



```python
hr_data = pd.read_csv('HR_comma_sep.csv.txt')
```

```python
hr_data.rename(columns={'sales':'dept'}, inplace=True)
```

```python
hr_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_salary = LabelEncoder()
```

```python
le_salary.fit(hr_data.salary)
```

```python
hr_data['salary_en'] = le_salary.transform(hr_data.salary)
```

```python
le_dept = LabelEncoder()
```

```python
le_dept.fit(hr_data.dept)
```

```python
hr_data['dept_en'] = le_dept.transform(hr_data.dept)
```

```python
hr_data.head()
```

```python
num_hr_data = hr_data.drop(columns=['salary','dept']) 
kmeans = KMeans(n_clusters= 3)
kmeans.fit(num_hr_data)
```

```python
labels = kmeans.predict(num_hr_data)
```

```python
labels
```

```python
num_hr_data[labels == 2]
```

```python
hr_data_subset = num_hr_data[['number_project','time_spend_company']]
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(hr_data_subset)
```

```python
labels = kmeans.predict(hr_data_subset)
```

```python
plt.scatter(x=hr_data_subset.time_spend_company, y=hr_data_subset.number_project, s=10, c=labels)
plt.xlabel('Time spend In Company')
plt.ylabel('No, of projects')
```
