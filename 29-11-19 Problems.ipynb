

Problem 1 In a normal supermarket, things are identified using Stock Keeping Units,. In our store, we’ll use individual letters of the alphabet (A, B, C, and so on). Our goods are priced individually. In addition, some items are multipriced: buy n of them, and they’ll cost you y Rs. For example, item ‘A’ might cost 50 Rs individually, but this week we have a special offer: buy three ‘A’s and they’ll cost you 120Rs. In fact this week’s prices are:

Item Unit Special Price Price A 50 3 for 120 B 30 2 for 45 C 20 D 15

Our checkout accepts items in any order, so that if we scan a B, an A, and another B, we’ll recognize the two B’s and price them at 45 . Because the pricing changes frequently, we need to be able to pass in a set of pricing rules each time we start handling a checkout transaction.


```python

class Item:
    items = []
    item_names = []
    @classmethod
    def add_ites(cls , obj):
        cls.items.append(obj)
        cls.item_names.append(obj.item)
    
    def __init__(self , item , price , count):
        self.item  = item 
        self.__price = price
        self.count = count
        Item.add_ites(self)
        
    def get_Price(self):
        return self.__price
    
    def set_Count(self , count):
        self.count = count
        
    def set_Price(self , price):
        self.__price = price
    
class Offers(Item):
    
    offers = []
    offer_item_names = []
    @classmethod
    def add_offers(cls , obj):
        cls.offers.append(obj)
        cls.offer_item_names.append(obj.item)
        
    def __init__(self , item , quant , price):
        if item in Item.item_names:
            self.item = item 
            self.quant = quant
            self.__price = price
            Offers.add_offers(self)
            
    def get_Price(self):
        return self.__price
    
    def get_quant(self):
        return self.quant
    

class Supermarket:
    
    @classmethod
    def calc_bill(cls,cart):
        total_bill = 0
        set_Of_Items = set(cart) 
        
        for item in set_Of_Items:
            count = cart.count(item)
            in_Items = Item.item_names.index(item)

            if item in Offers.offer_item_names:
                in_Offers = Offers.offer_item_names.index(item)

                if count >= Offers.offers[in_Offers].quant:
                    total_bill += (((count // Offers.offers[in_Offers].quant)* Offers.offers[in_Offers].get_Price()) + ((count % Offers.offers[in_Offers].quant)* Item.items[in_Items].get_Price()))
                else:
                    total_bill += count * Item.items[in_Items].get_Price()
                    
            else:
                total_bill += count * Item.items[in_Items].get_Price()                
                    
            
        return total_bill 
            


class Customer:
    
    customers = []
    
    @classmethod
    def add_customers(cls , obj):
        cls.customers.append(obj)
        
    def __init__(self, name , email):
        self.name = name
        self.email = email
        self.cart = []
        Customer.add_customers(self)
        
    def add_product_cart(self ,item):
        self.cart.append(item)
        
    def rm_product_cart(self ,item):
        if item in self.cart:
            self.cart.remove(item)
        else:
            print(item," is Not in the cart.")
            
        
    def view_cart(self):
        return self.cart
    
    def request_bill(self,cart):
        return Supermarket.calc_bill(cart)
    
    
class Admin(Offers):
    
    
    def __init__(self,un,pswd):
        self.user_name = un
        self.pswd = pswd
    
    def add_product(self):
        item_name = input("Enter the Item Name :")
        item_price = int(input("Enter the Item per unit price :"))
        item_count = int(input("Enter the count :"))
        items = Item(item_name,item_price,item_count)
        print("Sucessfully added an Item")
    
    def update_unit_price(self):
        item = input("\nEnter the Item name you want to update price :")
        new_price = int(input("\nEnter the new amount :"))
        in_Items = Item.item_names.index(item)
        Item.items[in_Items].set_Price(new_price)
        print("Sucessfully updated.")
        
    def view_all_products(self):
        print("Products List")
        for item in Item.items:
            print(item.__dict__)
        

itemA = Item("A",50,150)
itemB = Item("B",30,150)
itemC = Item("C",20,70)
itemD = Item("D",15,80)

offer1 = Offers("A",3,120)
offer2 = Offers("B",2,45)
```
