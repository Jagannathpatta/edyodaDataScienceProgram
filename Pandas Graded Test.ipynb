### 3. Find all houses which has more than 3 bedroom and price less than 50k. Update their price with 60k

```python
Q3DF = house_rental[(house_rental.Bedroom > 3) & (house_rental.Price < 50000)]
Q3DF.head()
```

### 1. Display all the information respective to gender.

```python
hc_df_gb_gender = hc_df.groupby(['Gender'])
```

### 10. Change place to urban if gender is female else rural.

```python
Q10DF = hc_df[(hc_df['Place(location where the patient lives)'] == 'rural')& (hc_df.Gender == 'female')]
print(Q10DF.shape)
Q10DF.head()
```

```python
def f(f):
    if (f.Gender == 'female'):
        return 'urban'
    else:
        return f['Place(location where the patient lives)']
hc_df['Place(location where the patient lives)'] = hc_df.apply(f, axis=1)
```
